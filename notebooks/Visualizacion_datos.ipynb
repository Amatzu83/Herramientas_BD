```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Scatter plot (diagramas de dispersion)
x = np.arange(1,20)
y = x**2
y1=x**2 + 5*x
plt.scatter(x,y,color='red',marker='s',label='x**2')
plt.scatter(x,y1,color='blue',marker='s',label='x**3+5x')
plt.legend(title='Funciones')
plt.title('tipo de finciones',fontsize=20)
plt.xlabel('x',color='green')
plt.ylabel('f(x)')
plt.grid()
plt.text(5,300,'estetexto')
#for val_x in x:
 #       plt.text(val_x,val_x**2,f
```
